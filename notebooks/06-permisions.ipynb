```python
%sql
CREATE SCHEMA IF NOT EXISTS security;

CREATE OR REPLACE FUNCTION security.filter_ca_rows(state_value STRING)
RETURNS BOOLEAN
RETURN
  is_member('admins')  -- admins see all rows
  OR ( is_member('cuenca') AND state_value = 'Cuenca' )
  OR ( is_member('not_cuenca') AND state_value <> 'Cuenca' );
```

```python
%sql
SElect distinct city FROM final_project.gold_daily_store_family SORT BY city LIMIT 5;
```

```python
%sql
ALTER TABLE cscie103_catalog.final_project.gold_daily_store_family
SET ROW FILTER cscie103_catalog.security.filter_ca_rows
ON (city);


ALTER TABLE cscie103_catalog.final_project.platinum_city_cagr
SET ROW FILTER cscie103_catalog.security.filter_ca_rows
ON (city);
```

```python
%sql
-- Grant catalog and schema access to both groups
GRANT USE CATALOG ON CATALOG cscie103_catalog TO `cuenca`;
GRANT USE CATALOG ON CATALOG cscie103_catalog TO `not_cuenca`;

GRANT USE SCHEMA ON SCHEMA cscie103_catalog.final_project TO `cuenca`;
GRANT USE SCHEMA ON SCHEMA cscie103_catalog.final_project TO `not_cuenca`;

-- Grant SELECT on all tables in the schema (simpler approach)
GRANT SELECT ON SCHEMA cscie103_catalog.final_project TO `cuenca`;
GRANT SELECT ON SCHEMA cscie103_catalog.final_project TO `not_cuenca`;
```
